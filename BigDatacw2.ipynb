```python
from pyspark.sql import SparkSession
import pandas as pd
# Create a Spark session
spark = SparkSession.builder \
        .appName("csv-analysis") \
        .getOrCreate()

# Read the CSV file into a DataFrame
df1 = spark.read.csv("IEA_Global_EV_Data2023.csv", header=True)

# Save the DataFrame to a CSV file with a header
df1.write.format("csv").option("header", "true").save("dfIEA_Global_EV_Data2023.csv")
# Count the number of rows in the DataFrame
row_count = df1.count()
print("Number of rows: ", row_count)

# Show the DataFrame
df1.show()
```

```python
df1.groupBy("parameter").count().show()
```

```python
df1.groupBy("powertrain").count().show()
```

```python
df1.groupBy("mode").count().show()
```

```python
df1.groupBy("region").count().show()
```

```python
# Filter the DataFrame for the specified regions and display the counts
df1.filter((df1["region"] == "Australia") |
           (df1["region"] == "Austria") |
           (df1["region"] == "Canada") |
           (df1["region"] == "Belgium") |
           (df1["region"] == "India") |
           (df1["region"] == "Germany") |
           (df1["region"] == "Japan") |
           (df1["region"] == "New Zealand") |
           (df1["region"] == "South Africa") |
           (df1["region"] == "United Kingdom")).groupBy("region").count().show()
```

```python
from pyspark.sql import SparkSession

# Create a Spark session
spark = SparkSession.builder \
        .appName("csv-analysis") \
        .getOrCreate()

# Read the CSV file into a DataFrame
df2 = spark.read.csv("Environment_Temperature_change_E_All_Data_NOFLAG.csv", header=True)

# Save the DataFrame to a CSV file with a header
df2.write.format("csv").option("header", "true").save("df2Environment_Temperature_change_E_All_Data_NOFLAG.csv")
# Count the number of rows in the DataFrame
row_count = df2.count()
print("Number of rows: ", row_count)

# Show the DataFrame
df2.show()
```

```python
# Filter the DataFrame for the specified regions and display the counts
df2.filter((df2["Area"] == "Australia") |
           (df2["Area"] == "Austria") |
           (df2["Area"] == "Canada") |
           (df2["Area"] == "Belgium") |
           (df2["Area"] == "India") |
           (df2["Area"] == "Germany") |
           (df2["Area"] == "Japan") |
           (df2["Area"] == "New Zealand") |
           (df2["Area"] == "South Africa") |
           (df2["Area"] == "United Kingdom")).groupBy("Area").count().show()
```
